```python
from PIL import Image, ImageDraw, ImageFont, ImageOps
from bidi.algorithm import get_display
import arabic_reshaper
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1.read color image
img = Image.open('Inputs/python.jpg')
rows, cols = img.size
```

```python
#2.write a Persian text
font = ImageFont.truetype('Inputs\MSUIGHUB.TTF', 300)
drawer = ImageDraw.Draw(img)

text = "پایتون"
reshaped_text = arabic_reshaper.reshape(text)
bidi_text = get_display(reshaped_text) 
drawer.text((rows//2, cols//2), bidi_text, fill=(205, 205, 0), font=font)
drawer = ImageDraw.Draw(img)


img.save('Outputs/Persian_text.jpg')
plt.imshow(img)
```

```python
#calculate color image histogram
r, g, b = img.split()

hr = r.histogram()
hg = g.histogram()
hb = b.histogram()

plt.plot(hr, 'red')
plt.plot(hg, 'green')
plt.plot(hb, 'blue')

plt.show()
```

```python
#4.ecualize color image histogram
equalized_img = ImageOps.equalize(img)

plt.imshow(equalized_img)
```

```python
#convert color image to grayscale
gray_img = img.convert('L')

plt.imshow(gray_img, cmap='gray')
```

```python
#calculate histogram of grayscale image
hist = gray_img.histogram()

plt.plot(hist, color='gray')

plt.show()
```

```python
#equalize histogram of grayscale image
equalized_gray_img = ImageOps.equalize(gray_img)

plt.imshow(equalized_gray_img, cmap='gray')
```
