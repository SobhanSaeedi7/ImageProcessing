```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("Inputs/Unequalized_Hawkes_Bay_NZ.jpg", 0)
equalize1 = cv2.equalizeHist(img1)
cl1 = cv2.createCLAHE(0)
clahe1 = cl1.apply(img1)
result1 = np.hstack((img1, equalize1, clahe1))
cv2.imwrite("Outputs/histogram_equalization/equalize1.jpg", result1)
```

```python
hist1 = cv2.calcHist([img1], [0], None, [256], [0, 256])

plt.plot(hist1)
plt.show()

hist2 = cv2.calcHist([equalize1], [0], None, [256], [0, 256])

plt.plot(hist2)
plt.show()

hist3 = cv2.calcHist([clahe1], [0], None, [256], [0, 256])

plt.plot(hist3)
plt.show()
```
